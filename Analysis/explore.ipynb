```python
import pandas as pd
import json
import random as rd
```

```python
with open("/Users/rishabh.singh/Desktop/markdown_filter/filter/results/confluence_analysis_results_b.json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)
```

```python
gibberish = []
useful = []

for index, row in df.iterrows():
    row_dict = row.to_dict()   # convert Series â†’ dict
    if row_dict['decision'] == 'gibberish':
        gibberish.append(row_dict)
    else:
        useful.append(row_dict)

print(f"Gibberish = {len(gibberish)}")
print(f"Useful = {len(useful)}")

with open("/Users/rishabh.singh/Desktop/markdown_filter/Analysis/gibberish_b.json", "w") as f:
    json.dump(gibberish, f, indent=4)

with open("/Users/rishabh.singh/Desktop/markdown_filter/Analysis/useful_b.json", "w") as f:
    json.dump(useful, f, indent=4)
```
